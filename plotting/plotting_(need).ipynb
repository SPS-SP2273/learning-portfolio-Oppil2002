# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 1, 4, 9, 16, 25, 36]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

```python
plt.plot(x, y_3, label='Test', lw='5') 
# Just plots the points in a line

plt.xlabel('x-values')
plt.ylabel('y-values')
# Introduces titles of the axes

plt.title('X vs Y')
# Introduces the title of the graph

plt.grid(alpha=1)
# Introduces gridlines where alpha determines the transparency

plt.legend(loc='lower right')
# Introduces a legend where loc determines the location
```

### Things to note

```python
plt.plot(x, y_3, c='black', ls='dashed', lw=2)
```

## 1.2 Adding another plot

```python
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.plot(x, y_3, label='y=x^2', color='black') 
#The label determines the name of the line on the legend
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.errorbar(x, y_3, label='Y3 with errors', xerr=0.1, yerr=0.25)

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 1, 4, 9, 16, 25, 36]
```

```python
import numpy as np
```

```python
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)
np_y_3 = np.array(y_3)
```

```python
plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x2 = x**2+1                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, y_3, label='reference', c='black')
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

## We need another axis!

```python
x2 = x**2
sin_x = np.sin(x)
x = np.linspace(0, 6, 5100)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

### Things to note

# 3 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=300)
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

## 4.2 Styles

```python
plt.style.use('classic')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
