```python
# Defining the dictionary
convert_month = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

import os
```

```python
def convert_filename(filename):
    mmdd, city_ext = filename.split('_')
    month, day = mmdd.split('-')
    city, ext = os.path.splitext(city_ext)
    month = convert_month.get(month.lower())
    if month is None:
        return None
    new_filename = f'{month}-{day}-{city}{ext}'
    return new_filename
```

## Task 4

```python
import glob
```

```python
def convert_filename(filepath):
    directory, filename = os.path.split(filepath)
    if filename[:2].isdigit() and filename[2] == '-' and filename[5] == '-':
        return
    try:
        mmdd, city_ext = filename.split('_')
        month, day = mmdd.split('-')
        city, ext = os.path.splitext(city_ext)
        month = convert_month.get(month.lower())
        if month is None:
            return None
        new_filename = f'{month}-{day}-{city}{ext}'
        new_filepath = os.path.join(directory, new_filename)
        os.rename(filepath, new_filepath)
        print(f"Renamed '{filename}' to '{new_filename}'")
    except IndexError:
        print(f"Unexpected format: {filename}, file skipped")


dir_path = 'os-collaboration-exercise-data'

for filepath in glob.glob(os.path.join(dir_path, '*.txt')):
    convert_filename(filepath)
```
